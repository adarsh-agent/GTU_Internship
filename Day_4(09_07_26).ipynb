```python
import pandas as pd
import numpy as np

# Create DataFrame
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}

energy_df = pd.DataFrame(data)

# Create copies
energy_df_mean = energy_df.copy()
energy_df_ffill = energy_df.copy()
energy_df_bfill = energy_df.copy()

# ---------------- Mean Imputation ----------------
energy_df_mean["Energy Consumption (MWh)"] = energy_df_mean["Energy Consumption (MWh)"].fillna(
    energy_df_mean["Energy Consumption (MWh)"].mean()
)

energy_df_mean["Cost (Million $)"] = energy_df_mean["Cost (Million $)"].fillna(
    energy_df_mean["Cost (Million $)"].mean()
)

print("Data After Mean Imputation:")
print(energy_df_mean)

# ---------------- Forward Fill ----------------
print("\nBefore Forward Fill:")
print(energy_df_ffill)

forward_filled_df = energy_df_ffill.ffill()

print("\nData After Forward Fill:")
print(forward_filled_df)

# ---------------- Backward Fill ----------------
print("\nBefore Backward Fill:")
print(energy_df_bfill)

backward_filled_df = energy_df_bfill.bfill()

print("\nData After Backward Fill:")
print(backward_filled_df)
```

```python
#ToDay Task
import pandas as pd
import numpy as np
```

```python
# Import Dataset

df = pd.read_csv("/content/renewable_energy.csv")

print("Original Dataset\n")
print(df.to_string(index=True))
```

```python
# Relationship Between Columns
# -----------------------------------------------------

# Energy Consumption ↔ CO2 Reduction
# Positive relationship.
# More renewable energy generally reduces more CO2.

# Energy Consumption ↔ Cost
# Moderate positive relationship.
# Large projects generally cost more.

# Energy Source ↔ Cost
# Different technologies have different installation costs.

# Country ↔ Energy Source
# Certain countries prefer specific renewable resources.

# Country ↔ CO2 Reduction
# Depends on renewable energy adoption.
```

```python
# Check Null Values
# -----------------------------------------------------

print("\nNull Values")
print(df.isnull().sum())
```

```python

# Forward Fill
df_forward = df.ffill()

print("\n========== Dataset After Forward Fill ==========")
print(df_forward)

# Backward Fill
df_backward = df.bfill()

print("\n========== Dataset After Backward Fill ==========")
print(df_backward)
```

```python
# Handle Missing Values
# -----------------------------------------------------
# Numerical columns -> Replace with Mean

print("\nDataset Before Handling Null Values")
print(df)
df["Energy Consumption (MWh)"] = df["Energy Consumption (MWh)"].fillna(
    df["Energy Consumption (MWh)"].mean()
)

df["Cost (Million $)"] = df["Cost (Million $)"].fillna(
    df["Cost (Million $)"].mean()
)

df["CO2 Reduction (Tons)"] = df["CO2 Reduction (Tons)"].fillna(
    df["CO2 Reduction (Tons)"].mean()
)

print("\nDataset After Handling Null Values")
print(df)
```

```python
# Analysis
# -----------------------------------------------------

print("\n<--Average Energy Consumption-->")
print(df["Energy Consumption (MWh)"].mean())

print("\nAverage Cost")
print(df["Cost (Million $)"].mean())

print("\n<--Average CO2 Reduction-->")
print(df["CO2 Reduction (Tons)"].mean())

print("\nEnergy Source Frequency")
print(df["Energy Source"].value_counts())

print("\nCountry Frequency")
print(df["Country"].value_counts())
```

```python
# Correlation
# -----------------------------------------------------

print("\nCorrelation Matrix")
print(df.select_dtypes(include=np.number).corr())
```
